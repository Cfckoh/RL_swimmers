```python
import julia
jl = julia.Julia(compiled_modules=False)
from julia import Main
import numpy as  np
import plotly.graph_objects as go
```

```python
Main.include("abc_numerics.jl")
```

```python
A = 1.
B = 0.7
C = 0.43
NU = 0.99
kappa = 0.001
BETA = 0.1
PHI = 0.
```

```python
u=np.zeros(3)#np.random.random(3)
active_traj = []
passive_traj = []
for i in range(1000):
    loc_a = np.array2string(u, separator=",")
    loc_p = np.array2string(u, separator=",")
    a,p,penalty = Main.eval(f"envStep({A},{B},{C}, {PHI}, {NU},{kappa}, {BETA},{loc_a},{loc_p},0.05)")
    active_traj.append(a)
    passive_traj.append(p)
```

```python
traj = np.array(traj)
```

```python
fig = go.Figure(data=go.Scatter3d(
        x=traj[:,0],
        y=traj[:,1],
        z=traj[:,2],
        mode='lines',
        name = "trajectory"
      )
    )
```

### Now testing the environment wrapping the numerics

```python
from abc_env import ABC_env
```

```python
D = 0.06 # placeholder for now
env = ABC_env(A,B,C,0.2,BETA,kappa,D,NU)
env.deltaT = 0.05
```

```python
traj = []
while not env.isOver():
    traj.append(env.getState())
    env.step(0.)
env.reset()
traj = np.array(traj)
    
```

```python
fig = go.Figure(data=go.Scatter3d(
        x=traj[:,0],
        y=traj[:,1],
        z=traj[:,2],
        mode='lines',
        name = "trajectory"
      )
    )
fig.show()
```
