```python
import numpy as np
import sdeint
import gym
import numpy as np
from gym import spaces
import stable_baselines3
from stable_baselines3.common.env_checker import check_env
from stable_baselines3 import A2C, PPO
import matplotlib.pyplot as plt
from abc_env_stoch_gym import ABCflow_gym
```

```python
env = ABCflow_gym()
check_env(env)
```

```python
eps=100
env = ABCflow_gym()
env.deltaT= 0.05
env.limit=10.
time_steps = int(env.limit/env.deltaT)*eps

model = A2C('MlpPolicy', env, verbose=True)
model.learn(total_timesteps=time_steps)
model.save("saved_models/A2C")
```

```python
eps=100
env = ABCflow_gym()
env.deltaT= 0.05
env.limit=10.
time_steps = int(env.limit/env.deltaT)*eps

model = PPO('MlpPolicy', env, verbose=True)
model.learn(total_timesteps=time_steps)
model.save("saved_models/PPO")
```

```python
model2 = PPO('MlpPolicy', env, verbose=True)
model2.load("saved_models/A2C")
```

```python
num_eps=1
cum_rew=np.zeros(int(env.limit/env.deltaT))
for ep in range(2):
    obs = env.reset()
    done = False
    time_step=0
    episode_rew=0
    while not done:
        action, _states = model.predict(obs)
        #print(action)
        obs, reward, done, info = env.step(action)
        episode_rew += reward
        cum_rew[time_step] += episode_rew
        time_step+=1
plt.plot(np.arange(len(cum_rew))/(1/env.deltaT),cum_rew/num_eps)
```
