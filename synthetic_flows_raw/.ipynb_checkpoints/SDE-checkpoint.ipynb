```julia
using DifferentialEquations
using Plots
using StatsBase
using LinearAlgebra
using CurveFit
```

```julia
function f(u,p,t)
    # r1 = u1, r2 = u2, s11=u3, s12=u4 s21=u5, s22=u6
    return [
        u[3]*u[1] + u[4]*u[2] - PHI * u[1]
        u[5]*u[1] + u[6]*u[2] - PHI * u[2]
        - 1/tau * u[3]
        - 1/tau * u[4]
        - 1/tau * u[5]
        - 1/tau * u[6]
    ]
end


function g(u,p,t)
    [
        k 0 0 0 0
        0 k 0 0 0
        0 0 s 0 0 
        0 0 0 s sqrt(2)*s
        0 0 0 s -sqrt(2)*s
        0 0 -s 0 0
    ]
end
```

```julia
tau = 1.0
k = 0.1
D = 0.001
s = sqrt(D)
PHI=0.5
r_d = sqrt(k/D)
```

```julia
u0 = zeros(6)
tspan = (0.0, 5000.0)
prob = SDEProblem(f, g, u0, tspan, noise_rate_prototype = zeros(6, 5))
#res = solve(prob, saveat=0:0.1:100)
res = solve(prob)
```

```julia
u = reduce(hcat, res.u)
plot(res.t, u[2,:])
plot!(res.t, u[1,:])
```

```julia
i = 200
sigma =  [u[3,i] u[4,i]; u[5,i] u[6,i]]
eigen(sigma)
```

```julia
plot(res.t, u[4,:])
```

```julia
seperations = sum(x -> x^2, u[1:2,:]; dims=1)
seperations .= sqrt.(seperations)
```
